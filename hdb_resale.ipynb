```python
df_malls=pd.read_csv('shopping_mall_location.csv')
df_malls.head()
```

```python

nrst_mrt_arr = [];
nrst_mrt_dis_arr =[];
nrst_mall_arr=[];
nrst_mall_dis_arr=[];

for i, row in df.iterrows():
    Distance_mrt=200
    MRT=0
    Distance_mall=200
    Mall=0
    if df['longitude'].loc[i] != 0:
        # check the nearst mrt station
        for j, row in df_mrt.iterrows():
            if df_mrt['longitude'].loc[j] != 0:
                temp_dis = geopy.distance.distance((df['latitude'].loc[i],df['longitude'].loc[i]),(df_mrt['latitude'].loc[j],df_mrt['longitude'].loc[j])).km
                if temp_dis < Distance_mrt:
                    Distance_mrt = temp_dis
                    MRT = df_mrt['mrt_station'].loc[j]
                else:
                    continue
            else:
                continue
        # check the nearst shopping mall
        for j, row in df_malls.iterrows():
            if df_malls['longitude'].loc[j] != 0:
                temp_dis_mall = geopy.distance.distance((df['latitude'].loc[i],df['longitude'].loc[i]),(df_malls['latitude'].loc[j],df_malls['longitude'].loc[j])).km
                if temp_dis_mall < Distance_mall:
                    Distance_mall = temp_dis_mall
                    Mall = df_malls['shopping Mall'].loc[j]
                else:
                    continue
            else:
                continue
        
    else:
        continue

    nrst_mrt_arr = np.append(nrst_mrt_arr, MRT)
    nrst_mrt_dis_arr = np.append(nrst_mrt_dis_arr, Distance_mrt)
    nrst_mall_arr = np.append(nrst_mall_arr, Mall)
    nrst_mall_dis_arr = np.append(nrst_mall_dis_arr, Distance_mall)
    
    print(i,":",df['street_name'].loc[i]," clst_mrt-",nrst_mrt_arr[i]," dist-",nrst_mrt_dis_arr[i])
    print(i,":",df['street_name'].loc[i]," clst_mall-",nrst_mall_arr[i]," dist-",nrst_mall_dis_arr[i])
        
    
```

```python
print(nrst_mrt_arr.size)
df['nearst_mrt']=nrst_mrt_arr
df['distance_nearst_mrt']=nrst_mrt_dis_arr
df['nearst_mall']=nrst_mall_arr
df['distance_nearst_mall']=nrst_mall_dis_arr
df.head()
```
